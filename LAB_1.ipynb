## Task #2: Loading and Visualizing Images

```python
import cv2
import matplotlib.pyplot as plt

# Load the image data into a NumPy array
# Note: OpenCV reads images in BGR format by default. Matplotlib expects RGB.
# So, we convert it for correct display with matplotlib.
img_bgr = cv2.imread('/content/Cameramantif-Table-2-Results-Obtained-using-Baseline_Q320.jpg')

if img_bgr is None:
    print("Error: Could not load image 'images/cameraman.tif'. Please ensure the file exists and the path is correct.")
else:
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.title('Image loaded with OpenCV')
    plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Load image into a PIL Image object
try:
    img2 = Image.open('/content/Cameramantif-Table-2-Results-Obtained-using-Baseline_Q320.jpg')
    plt.imshow(img2, cmap = cm.Greys_r)
    plt.title('Image loaded with PIL (Grayscale)')
    plt.show()
except FileNotFoundError:
    print("Error: Could not load image 'images/lena_gray_256.tif'. Please ensure the file exists and the path is correct.")
```

```python
import cv2

# Assuming 'img_bgr' from Task 2a is available and loaded successfully
if 'img_bgr' in locals() and img_bgr is not None:
    cv2.imwrite('/content/Cameramantif-Table-2-Results-Obtained-using-Baseline_Q320.jpg', img_bgr)
    print("Image saved as 'new_image_opencv.jpg' using OpenCV.")
else:
    print("Cannot save image with OpenCV: 'img_bgr' is not loaded. Please run Task 2a first.")
```

```python
import numpy as np

# Display shape and content of OpenCV loaded image (which is a NumPy array)
# Assuming 'img_bgr' from Task 2a is available and loaded successfully
if 'img_bgr' in locals() and img_bgr is not None:
    print("Shape of OpenCV image array (img_bgr):")
    print(img_bgr.shape)
    print("Content of OpenCV image array (first 5x5 pixels of first channel):")
    print(img_bgr[:5, :5, 0])
else:
    print("OpenCV image array 'img_bgr' is not loaded. Please run Task 2a first.")
```

```python
import numpy as np
from PIL import Image

# Convert the PIL image to a NumPy array and print its shape and content
# Assuming 'img2' from Task 2b is available and loaded successfully
if 'img2' in locals():
    img_array = np.array(img2)
    print("\nShape of PIL image array (img_array):")
    print(img_array.shape)
    print("Content of PIL image array (first 5x5 pixels):")
    print(img_array[:5, :5])
else:
    print("PIL image 'img2' is not loaded. Please run Task 2b first.")
```
